```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ---- Convert Dates ----
DF1['Date'] = pd.to_datetime(DF1['Date'], dayfirst=True, errors='coerce')
DF2['Date'] = pd.to_datetime(DF2['Date'], dayfirst=True, errors='coerce')
DF1 = DF1.dropna(subset=['Date'])
DF2 = DF2.dropna(subset=['Date'])

# ---- Merge parasite and hazard data on Date, Country, Region ----
df = pd.merge(DF1, DF2, on=['Date', 'Country', 'Region'], how='inner')

# ---- Convert numeric columns ----
df['RR'] = df['RR'].astype(str).str.replace(',', '.').astype(float)
df['Percentage_y'] = df['Percentage_y'].astype(str).str.replace(',', '.').astype(float)

# ---- Keep only rows where RR and Percentage_y are present ----
df = df.dropna(subset=['RR', 'Percentage_y'])

# ---- Calculate median for Percentage_y ----
median_percentage = df['Percentage_y'].median()

# ---- Identify co-variation categories ----
df.loc[(df['RR'] > 1.0) & (df['Percentage_y'] > median_percentage), 'CoVariation'] = 'High co-variation'
df.loc[(df['RR'] > 1.0) & (df['Percentage_y'] <= median_percentage), 'CoVariation'] = 'Moderate co-variation'
df.loc[(df['RR'] < 1.0) & (df['Percentage_y'] >= median_percentage), 'CoVariation'] = 'Low co-variation'
df.loc[(df['RR'] < 1.0) & (df['Percentage_y'] <= median_percentage), 'CoVariation'] = 'Very low co-variation'

# ---- Ensure totals only count valid rows ----
valid_df = df.dropna(subset=['RR', 'Percentage_y'])

# ---- Compute co-variation percentages per country ----
country_stats = valid_df.groupby(['Country', 'CoVariation']).size().reset_index(name='Count')
country_totals = valid_df.groupby('Country').size().reset_index(name='Total')
country_stats = country_stats.merge(country_totals, on='Country')
country_stats['Category_Percentage'] = (country_stats['Count'] / country_stats['Total']) * 100

print("Percentage of points in each co-variation category by country:")
print(country_stats)

# ---- Compute co-variation percentages per region ----
region_stats = valid_df.groupby(['Region', 'CoVariation']).size().reset_index(name='Count')
region_totals = valid_df.groupby('Region').size().reset_index(name='Total')
region_stats = region_stats.merge(region_totals, on='Region')
region_stats['Category_Percentage'] = (region_stats['Count'] / region_stats['Total']) * 100

print("\nPercentage of points in each co-variation category by region:")
print(region_stats)

# ---- Save country stats to CSV ----
country_stats.to_csv('country_covariation_stats.csv', index=False)

# ---- Save region stats to CSV ----
region_stats.to_csv('region_covariation_stats.csv', index=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# ---- Set a clean, publication-ready style with larger fonts ----
sns.set(style="whitegrid")
plt.rcParams.update({
    "font.family": "serif",
    "font.size": 30,            # general font size
    "axes.titlesize": 22,       # title size
    "axes.labelsize": 22,       # axis label size
    "xtick.labelsize": 22,      # x tick labels
    "ytick.labelsize": 22,      # y tick labels
    "legend.fontsize": 22,      # legend font
    "figure.figsize": (22, 12)
})

# ---- Nature journal-friendly, muted and colorblind-friendly palette ----
palette = ["#4E79A7", "#E15759","#F28E2B", "#76B7B2"]  # soft blue, orange, red, teal

# ---- Create the bar plot using the correct 'Category_Percentage' ----
plt.figure()
sns.barplot(
    data=region_stats,
    x='Region',
    y='Category_Percentage',  # corrected from 'Percentage_y'
    hue='CoVariation',
    palette=palette,
    edgecolor='black'
)

# ---- Aesthetic tweaks ----
plt.xticks( ha='right')
plt.ylabel('Percentage (%)', labelpad=22)
plt.xlabel('')
plt.title('Co-variation Categories by Region', pad=22)
plt.legend(title='Co-variation', frameon=True, framealpha=0.4, edgecolor='black')
sns.despine(trim=True)

plt.tight_layout()
plt.legend().remove()
plt.savefig('Figure_char_hazard.png', dpi=300)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ---- Convert Dates ----
DF1['Date'] = pd.to_datetime(DF1['Date'], dayfirst=True, errors='coerce')
DF4['Date'] = pd.to_datetime(DF4['Date'], dayfirst=True, errors='coerce')
DF1 = DF1.dropna(subset=['Date'])
DF4 = DF4.dropna(subset=['Date'])

# ---- Merge parasite and hazard data on Date and Country ----
df = pd.merge(DF1, DF4, on=['Date', 'Country', 'Region'], how='inner')

# ---- Convert numeric columns ----
df['RR'] = df['RR'].astype(str).str.replace(',', '.').astype(float)
df['Precipitation_mean__mm'] = df['Precipitation_mean__mm'].astype(str).str.replace(',', '.').astype(float)

# ---- Keep only rows with valid RR and Precipitation ----
df = df.dropna(subset=['RR', 'Precipitation_mean__mm'])

# ---- Calculate median for Precipitation ----
median_percentage = df['Precipitation_mean__mm'].median()

# ---- Identify co-variation categories ----
df.loc[(df['RR'] > 1.0) & (df['Precipitation_mean__mm'] > median_percentage), 'CoVariation'] = 'High co-variation'
df.loc[(df['RR'] > 1.0) & (df['Precipitation_mean__mm'] <= median_percentage), 'CoVariation'] = 'Moderate co-variation'
df.loc[(df['RR'] < 1.0) & (df['Precipitation_mean__mm'] >= median_percentage), 'CoVariation'] = 'Low co-variation'
df.loc[(df['RR'] < 1.0) & (df['Precipitation_mean__mm'] <= median_percentage), 'CoVariation'] = 'Very low co-variation'

# ---- Compute co-variation percentages per country ----
country_stats = df.groupby(['Country', 'CoVariation']).size().reset_index(name='Count')
country_totals = df.groupby('Country').size().reset_index(name='Total')
country_stats = country_stats.merge(country_totals, on='Country')
country_stats['Percentage'] = (country_stats['Count'] / country_stats['Total']) * 100

print("Percentage of points in each co-variation category by country:")
print(country_stats)

# ---- Compute co-variation percentages per region ----
region_stats = df.groupby(['Region', 'CoVariation']).size().reset_index(name='Count')
region_totals = df.groupby('Region').size().reset_index(name='Total')
region_stats = region_stats.merge(region_totals, on='Region')
region_stats['Percentage'] = (region_stats['Count'] / region_stats['Total']) * 100

print("\nPercentage of points in each co-variation category by region:")
print(region_stats)

# ---- Save country stats to CSV ----
country_stats.to_csv('country_covariation_stats.csv', index=False)

# ---- Save region stats to CSV ----
region_stats.to_csv('region_covariation_stats.csv', index=False)
```
