We calculate the Geometric mean based 
on the RR of each country

```python
import numpy as np
from scipy import stats

RR = [
    1.32756131060781,
    1.77464148269661,
    1.05573135915684,
    1.28225596521123,
    1.05470101174961,
    1.10676455967986,
    1.14007321469917,
    1.31388214231128,
    0.753377936307964,
    2.57321253173416,
    1.42745737693897,
    2.38705877077746,
    2.72257189193688e-05,
    1.36069116679506,
    0.718022847222855,
    0.851600627137108,
    0.838476441950651,
    4.80670037177805,
    1.00107710831293,
    1.07456438377453
]





# Log-transform
log_RR = np.log(RR)

# Geometric mean
mean_log = np.mean(log_RR)
GM = np.exp(mean_log)

# Standard deviation and SE
s = np.std(log_RR, ddof=1)
n = len(RR)
SE = s / np.sqrt(n)

# 95% CI
t_value = stats.t.ppf(0.975, df=n-1)
margin = t_value * SE
CI_lower = np.exp(mean_log - margin)
CI_upper = np.exp(mean_log + margin)

# Convert to percentage
GM_percent = (GM - 1) * 100
CI_lower_percent = (CI_lower - 1) * 100
CI_upper_percent = (CI_upper - 1) * 100

# p-value for H0: mean_log = 0
t_stat = mean_log / SE
p_value = 2 * (1 - stats.t.cdf(np.abs(t_stat), df=n-1))

print(f"Geometric Mean RR: {GM:.3f} ({GM_percent:.1f}%)")
print(f"95% CI: [{CI_lower:.3f}, {CI_upper:.3f}] [{CI_lower_percent:.1f}%, {CI_upper_percent:.1f}%]")
print(f"t-statistic: {t_stat:.3f}")
print(f"p-value: {p_value:.4f}")
```
